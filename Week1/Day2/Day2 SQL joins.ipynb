```sql
select e.emp_name,emp.emp_name from employees e left join employees emp on e.manager_id = emp.emp_id;
```

```sql
select e.emp_name,d.dept_name from employees e join departments d on e.dept_id=d.dept_id;
```

```sql
select e.emp_name,d.emp_name from employees e inner join employees d on e.manager_id = d.emp_id;
```

```sql
select e.emp_name,d.dept_name from employees e left join departments d on e.dept_id=d.dept_id where d.dept_id is null;
```

```sql
select e.emp_name,p.project_name from employees e left join projects p on e.emp_id = p.emp_id;
```

```sql
select e.emp_name,p.project_name from employees e inner join projects p on e.emp_id = p.emp_id;
```

```sql
select p.project_name, e.emp_name from employees e right join projects p on p.emp_id = e.emp_id;
```

```sql
select e.emp_name,d.dept_name  from employees e left join departments d on e.dept_id=d.dept_id;
```

```sql
select d.dept_name,e.emp_name  from employees e right join departments d on e.dept_id=d.dept_id;
```

```sql
select e.emp_name,d.dept_name from employees e full outer join departments d on e.dept_id=d.dept_id; 
```
