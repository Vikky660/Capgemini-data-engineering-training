```sql
select *, CASE 
when experience >= '8' AND performance_rating = 'A' then '20%'
when experience >= '5' and performance_rating = 'B' then '15%'
when performance_rating = 'C' then 'no'
else '10%' 
end as `hike` from Employee;
```

```sql
select *, CASE 
when department='finance' then case
WHEN performance_rating = 'A' then '20%'
WHEN performance_rating = 'B' then '15%'
WHEN performance_rating = 'C' then '5%'
end 
when department='Engineering' then case
when performance_rating = 'A' then '18%'
WHEN performance_rating = 'B' then '12%'
WHEN performance_rating = 'C' then '3%'
end 
else '10%'
end as bonus from Employee
```

```sql
select *, case 
when salary > 80000 and performance_rating = 'A' then 'High'
when salary > 50000 and performance_rating = 'B' then 'Mid'
when salary < 50000 or performance_rating ='C' then 'Low'
end as performer from Employee
```

```sql
select *,
case 
when department ='HR' then case 
when experience < 5 then 'High'
when experience > 5 then 'Low'
end 
when department in ('Engineering','Finance') then
case
when experience > 8 then 'low'
when experience < 8 then 'medium'
end
else 'medium'
end as Risk from Employee
```

Nested Case When

```sql
select *,
case 
when performance_rating = 'A' then 
case 
when Salary >80000 and experience > 5 then '25%'
when Salary between 50000 and 80000 then '15%'
else '20%'
end 
when performance_rating = 'B' then
case
when experience > 5 then '12%'
else '10%'
end
when performance_rating = 'C' then 'No hike'
end as hike from employee
```

```sql
select * ,
case 
when department = 'Finance' then 
case 
when performance_rating ='A' and experience > 10 then '25%'
else '20%'
end 
when department = 'HR' then 
case 
when performance_rating = 'B' and experience > 5 then '15%'
else '10%'
end 
else '8%'
end as bonus from Employee
```

```sql
select *,
case 
when Salary> 70000 then 
CASE  
when performance_rating ='A' and experience > 8 then 'top performer'
when experience < 8 then 'Mid performer'
end 

when salary between 50000 and 70000 and performance_rating = 'A' then 'Rising Star'
when Salary < 50000 then 'low performer'
end as performer from employee;
```

```sql
select *,
case 
when salary >90000 then 
case  
when experience >10 then '35%'
else '30%'
end 
when salary between 60000 and 90000 then 
case 
when experience >5 then '25%'
else '20%'
end 
when salary <50000 then '15%'
end as tax from employee 
```

```sql
select * ,
case 
when performance_rating = 'A' then 
case 
when salary >75000 and experience >7 then 'Eligible for Senior Role'
else 'Eligible for junior role'
end 
when performance_rating = 'B' then 
case 
when experience >5 then 'Eligible Concideration'
else 'Not eligible'
end 
when performance_rating ='C' then 'Not eligible'
end as Promotion from employee
```
