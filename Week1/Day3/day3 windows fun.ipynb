```sql
select *, row_number() over(order by salary desc) as row_num from emp1;
```

```sql
select * , row_number() over (partition by department order by salary desc) as row_num from emp1;
```

```sql
select * , row_number() over (order by join_date) as row_num from emp1;
```

```sql
select *, row_number() over (partition by department order by join_date) as row_num from emp1
```

```sql
select * , row_number() over (order by salary asc) as row_num from emp1;
```

```sql
select * , row_number() over(partition by department order by emp_name asc) as row_num from emp1;
```
